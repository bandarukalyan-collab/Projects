# Python Iterators and Python Internals

## 1. Introduction to Iterators

### What is an Iterator?
- An iterator is an object that implements the iterator protocol
- It allows you to traverse through all the elements of a collection
- Iterators are stateful - they remember their current position
- They provide a way to access elements one at a time without loading everything into memory

### Iterator Protocol
Python's iterator protocol consists of two methods:
- `__iter__()`: Returns the iterator object itself
- `__next__()`: Returns the next item from the container
- When there are no more items, `__next__()` raises `StopIteration`

```python
# Iterable vs Iterator example
my_list = [1, 2, 3, 4, 5]

# Iterable - can be iterated multiple times
for item in my_list:
    print(item)

for item in my_list:
    print(item)  # Works again

# Iterator - can only be iterated once
my_iterator = iter(my_list)
print(next(my_iterator))  # 1
print(next(my_iterator))  # 2
print(next(my_iterator))  # 3
```

## 3. Creating Custom Iterators

### Implementing __iter__ and __next__

To create a custom iterator, you need to implement:
- `__iter__()`: Returns the iterator object (usually self)
- `__next__()`: Returns the next value or raises StopIteration

```python
# Custom iterator example
class CountDown:
    def __init__(self, start):
        self.current = start
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current <= 0:
            raise StopIteration
        value = self.current
        self.current -= 1
        return value

# Using the custom iterator
countdown = CountDown(5)
for number in countdown:
    print(number)
```

## 4. Python Internals

### Object Model
- Everything in Python is an object
- Objects have identity, type, and value
- Objects are allocated on the heap
- Reference counting is used for memory management

### Method Resolution Order (MRO)
- Determines the order in which methods are searched in inheritance
- Used in multiple inheritance
- Follows C3 linearization algorithm
- Can be viewed using `ClassName.__mro__` or `ClassName.mro()`

```python
# MRO example
class A:
    def method(self):
        print("A")

class B(A):
    def method(self):
        print("B")

class C(A):
    def method(self):
        print("C")

class D(B, C):
    pass

d = D()
d.method()  # Which method is called?

# View MRO
print(D.__mro__)
```

## 5. Generators as Iterators

### Generator Functions
- Functions that use `yield` keyword
- Automatically implement the iterator protocol
- Maintain state between yields
- Memory efficient for large sequences

```python
# Generator as iterator
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

# Using the generator
for number in count_up_to(5):
    print(number)
```

## 6. Key Points & Summary

### Iterator Protocol
- `__iter__()`: Returns iterator object
- `__next__()`: Returns next value or raises StopIteration
- Iterators maintain state
- Can only be traversed once

### Iterable vs Iterator
- Iterable: Can be iterated multiple times
- Iterator: Can only be iterated once
- Every iterator is iterable
- Not every iterable is an iterator

### Python Internals
- Everything is an object
- Reference counting for memory management
- MRO determines method resolution in inheritance
- Generators automatically implement iterator protocol

### When to Use Custom Iterators
- When you need custom iteration logic
- When working with large datasets (memory efficiency)
- When you need to maintain state during iteration
- When implementing lazy evaluation
