# Python Memory Management


## 1. Introduction

Memory management means how Python stores, uses, and releases memory while a program is running.

Python handles most memory management automatically, so developers do not need to manually allocate and free memory like in some other languages.


## 2. Objects and References

In Python, everything is an object. Variables are references to objects, not direct containers of values.


````python
a = [1, 2, 3]
b = a

print(id(a))
print(id(b))

b.append(4)
print(a)
print(b)
````

## 3. Reference Counting

Python keeps track of how many references point to an object. When the reference count becomes zero, the object can be removed from memory.


````python
import sys

numbers = [1, 2, 3]
print(sys.getrefcount(numbers))

another_reference = numbers
print(sys.getrefcount(numbers))
````

## 5. Circular References

### What are Circular References?
- When two or more objects reference each other
- Reference counting cannot clean them up
- Garbage collector handles these cases

### Example of Circular Reference
```python
a = []
b = []
a.append(b)
b.append(a)
# Now a and b reference each other
```

````python
# Circular reference example
import gc

# Create circular reference
a = []
b = []
a.append(b)
b.append(a)

# Check reference counts
print(f"Ref count of a: {sys.getrefcount(a)}")
print(f"Ref count of b: {sys.getrefcount(b)}")

# Delete references
del a
del b

# Force garbage collection
gc.collect()
print("Garbage collection completed")
````

## 7. Memory Leaks

### Common Causes of Memory Leaks
- Circular references
- Global variables that keep growing
- Caching without size limits
- Unclosed file handles or connections
- Event listeners not removed

### Detecting Memory Leaks
- Use `gc.get_objects()` to see all objects
- Monitor memory usage over time
- Use profiling tools like `memory_profiler`

````python
# Detecting memory leaks
import gc
import sys

# Count objects by type
def count_objects():
    objects = gc.get_objects()
    type_counts = {}
    for obj in objects:
        obj_type = type(obj).__name__
        type_counts[obj_type] = type_counts.get(obj_type, 0) + 1
    
    # Show top 10 most common types
    sorted_types = sorted(type_counts.items(), key=lambda x: x[1], reverse=True)[:10]
    for obj_type, count in sorted_types:
        print(f"{obj_type}: {count}")

count_objects()
````

## 8. Memory Optimization Techniques

### 1. Use Generators Instead of Lists
- Generators yield values one at a time
- Much more memory efficient for large datasets
- Use `yield` instead of building full lists

### 2. Use __slots__
- Reduces memory usage for classes
- Prevents dynamic attribute creation
- Useful for classes with many instances

### 3. Use Data Classes with frozen=True
- Immutable data structures
- More memory efficient than regular classes

### 4. Use Array Module or NumPy
- For numerical data, use specialized arrays
- More memory efficient than Python lists

````python
# __slots__ example
class RegularClass:
    def __init__(self, x, y):
        self.x = x
        self.y = y

class SlotClass:
    __slots__ = ['x', 'y']
    def __init__(self, x, y):
        self.x = x
        self.y = y

# Compare memory usage
import sys
regular = RegularClass(1, 2)
slot = SlotClass(1, 2)

print(f"Regular class size: {sys.getsizeof(regular)}")
print(f"Slot class size: {sys.getsizeof(slot)}")

# With __slots__, you cannot add new attributes
try:
    slot.z = 3
except AttributeError as e:
    print(f"Cannot add new attribute: {e}")
````
