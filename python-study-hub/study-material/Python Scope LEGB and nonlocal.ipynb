# Python Scope, LEGB Rule and nonlocal Keyword


## 1. Introduction to Scope

Scope defines where a variable can be accessed in a program. In Python, a variable created in one place may not be available everywhere.

Understanding scope helps avoid confusion when the same variable name is used in different places.


```python
x = "global variable"

def display():
    x = "local variable"
    print("Inside function:", x)

display()
print("Outside function:", x)
```

## 2. LEGB Rule

Python searches for variables using the LEGB rule:

| Order | Scope | Meaning |
| --- | --- | --- |
| L | Local | Inside the current function |
| E | Enclosing | Inside outer function when functions are nested |
| G | Global | At module/file level |
| B | Built-in | Predefined names in Python |


```python
name = "global"

def outer():
    name = "enclosing"

    def inner():
        name = "local"
        print(name)

    inner()

outer()
```

## 3. global Keyword

The `global` keyword is used when we want to modify a global variable inside a function.


```python
counter = 0

def increment():
    global counter
    counter += 1

increment()
increment()
print(counter)
```

## 4. nonlocal Keyword

The `nonlocal` keyword is used inside a nested function to modify a variable from the enclosing function.

It is not used for global variables. It is used only when there is an outer function and an inner function.


```python
def outer():
    count = 0

    def inner():
        nonlocal count
        count += 1
        print(count)

    inner()
    inner()

outer()
```

## 5. Practical Examples of LEGB Rule

### Example 1: Nested Functions and Enclosing Scope

In nested functions, Python can access variables from the enclosing function scope.


```python
def outer():
    enclosing_var = "enclosing"

    def middle():
        local_var = "local"

        def inner():
            # LEGB search: local -> enclosing -> global -> built-in
            print(f"Accessing enclosing: {enclosing_var}")
            print(f"Accessing local: {local_var}")

        inner()

    middle()

outer()
```

### Example 2: Modifying Global Variables

Use `global` only when a function really needs to update a variable from the module level.


```python
count = 0  # Global variable

def increment():
    global count
    count += 1
    print(f"Inside function: {count}")

increment()
increment()
print(f"Outside function: {count}")
```

### Example 3: nonlocal with Counter

`nonlocal` is useful when nested functions need to maintain or update state from the outer function.


```python
def make_counter():
    count = 0  # Enclosing variable

    def increment():
        nonlocal count
        count += 1
        return count

    def decrement():
        nonlocal count
        count -= 1
        return count

    return increment, decrement

inc, dec = make_counter()
print(f"Increment: {inc()}")  # 1
print(f"Increment: {inc()}")  # 2
print(f"Decrement: {dec()}")  # 1
```

### Additional Example: Closure with nonlocal

This example shows how an inner function can remember and update a value from the enclosing scope.


```python
def make_multiplier(factor):
    def multiply(number):
        nonlocal factor
        result = number * factor
        factor += 1
        return result, factor

    return multiply

times_three = make_multiplier(3)
print(times_three(5))  # (15, 4)
print(times_three(5))  # (20, 5)
```

## Key Points

```python
- Scope controls where variables can be accessed.
- Python follows the LEGB rule to search variable names: Local -> Enclosing -> Global -> Built-in.
- `global` is used to modify global variables inside a function.
- `nonlocal` is used to modify enclosing function variables, not global variables.
- Avoid unnecessary global variables because they can make code harder to maintain.
- Use closures and `nonlocal` for maintaining state in nested functions.
- Understanding scope is important for writing clean, bug-free code.
```
