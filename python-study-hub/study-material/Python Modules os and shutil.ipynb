# Python Modules: os and shutil

## 1. Introduction to os Module

### What is the os Module?
- Provides functions for interacting with the operating system
- Allows you to work with files and directories
- Provides access to environment variables
- Platform-independent way to handle file system operations

### When to Use os Module
- Creating, reading, and deleting directories
- Working with file paths
- Getting system information
- Managing environment variables

## 2. Common os Module Functions

### Working with Directories
- `os.getcwd()`: Get current working directory
- `os.chdir(path)`: Change current working directory
- `os.listdir(path)`: List files and directories
- `os.mkdir(path)`: Create a directory
- `os.makedirs(path)`: Create directories recursively
- `os.rmdir(path)`: Remove a directory
- `os.removedirs(path)`: Remove directories recursively

```python
# os module examples
import os

# Get current working directory
print(f"Current directory: {os.getcwd()}")

# List files in current directory
print(f"Files in directory: {os.listdir()}")

# Create a directory
os.makedirs("test_dir", exist_ok=True)
print("Directory created")

# Change directory
os.chdir("test_dir")
print(f"Changed to: {os.getcwd()}")

# Go back to parent directory
os.chdir("..")
print(f"Back to: {os.getcwd()}")
```

## 3. Working with File Paths

### os.path Module
- `os.path.join()`: Join path components intelligently
- `os.path.abspath()`: Get absolute path
- `os.path.exists()`: Check if path exists
- `os.path.isfile()`: Check if path is a file
- `os.path.isdir()`: Check if path is a directory
- `os.path.basename()`: Get the filename
- `os.path.dirname()`: Get the directory name
- `os.path.splitext()`: Split filename and extension

```python
# os.path examples
import os

# Join paths (cross-platform)
path = os.path.join("folder", "subfolder", "file.txt")
print(f"Joined path: {path}")

# Get absolute path
abs_path = os.path.abspath("test.py")
print(f"Absolute path: {abs_path}")

# Check if path exists
print(f"Path exists: {os.path.exists('test_dir')}")

# Check if file or directory
print(f"Is file: {os.path.isfile('test_dir')}")
print(f"Is directory: {os.path.isdir('test_dir')}")

# Split filename and extension
filename, extension = os.path.splitext("document.pdf")
print(f"Filename: {filename}, Extension: {extension}")
```

## 4. Introduction to shutil Module

### What is the shutil Module?
- High-level file operations
- Built on top of os module
- Provides easier-to-use functions for file operations
- Cross-platform file copying and moving

### When to Use shutil Module
- Copying files and directories
- Moving files and directories
- Removing directory trees
- Archiving operations

```python
# shutil module examples
import shutil
import os

# Create test file
with open("source.txt", "w") as f:
    f.write("Test content")

# Copy file
shutil.copy("source.txt", "copy.txt")
print("File copied")

# Copy file with metadata
shutil.copy2("source.txt", "copy2.txt")
print("File copied with metadata")

# Move/rename file
shutil.move("source.txt", "moved.txt")
print("File moved")

# Create directory tree
os.makedirs("tree/subtree", exist_ok=True)

# Copy directory tree
shutil.copytree("tree", "tree_copy")
print("Directory tree copied")
```

## 5. File and Directory Deletion

### shutil.rmtree()
- Remove entire directory tree
- Useful for cleaning up temporary directories
- Cannot recover deleted files

### os.remove()
- Remove a single file
- Cannot remove directories

```python
# Deletion examples
import shutil
import os

# Remove a single file
if os.path.exists("moved.txt"):
    os.remove("moved.txt")
    print("File removed")

# Remove directory tree
if os.path.exists("tree_copy"):
    shutil.rmtree("tree_copy")
    print("Directory tree removed")

# Clean up test directory
if os.path.exists("test_dir"):
    shutil.rmtree("test_dir")
    print("Test directory removed")
```

## 6. Environment Variables

### os.environ
- Dictionary-like object for environment variables
- Access environment variables
- Set environment variables
- Used for configuration and secrets

```python
# Environment variables examples
import os

# Get environment variable
home = os.environ.get("HOME")
print(f"Home directory: {home}")

# Get with default
path = os.environ.get("MY_VAR", "default_value")
print(f"My variable: {path}")

# Set environment variable
os.environ["MY_VAR"] = "my_value"
print(f"Set MY_VAR: {os.environ['MY_VAR']}")

# List all environment variables
print("All environment variables:")
for key, value in os.environ.items():
    print(f"{key}: {value}")
```

## 7. Key Points & Summary

### os Module
- Low-level operating system interface
- Working with directories and files
- Path manipulation with os.path
- Environment variable access

### shutil Module
- High-level file operations
- Copy, move, and delete files/directories
- Directory tree operations
- Cross-platform compatibility

### Best Practices
- Use os.path.join() for cross-platform path handling
- Use shutil for high-level file operations
- Always check if paths exist before operations
- Use context managers for file operations
- Clean up temporary directories

### Common Use Cases
- Batch file processing
- Directory organization
- Backup and restore operations
- System administration scripts
- Data pipeline operations
