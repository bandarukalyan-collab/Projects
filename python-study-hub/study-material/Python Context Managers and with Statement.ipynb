# Python Context Managers and with Statement


## 1. Introduction to Context Managers

A context manager is used to manage resources properly. It automatically handles setup and cleanup operations.

Common examples:

- Opening and closing files
- Database connections
- Locks in threading
- Network connections

The `with` statement is used to work with context managers.


## 2. File Handling Without with

If we open a file manually, we must remember to close it.


```python
file = open("sample.txt", "w")
file.write("Hello Python")
file.close()
```

## 3. File Handling With with Statement

When we use `with`, Python automatically closes the file after the block is completed.


```python
with open("sample.txt", "w") as file:
    file.write("Hello Python using with statement")

print("File operation completed")
```

## 5. contextlib Module

### @contextmanager Decorator
- Simpler way to create context managers
- Uses generator function with yield
- Automatically handles cleanup in finally block

```python
# contextlib example
from contextlib import contextmanager

@contextmanager
def simple_context():
    print("Entering context")
    try:
        yield "Resource"
    finally:
        print("Cleaning up")

with simple_context() as resource:
    print(f"Using: {resource}")
```

## 6. Exception Handling in Context Managers

### Handling Exceptions in __exit__
- `__exit__` receives exception information
- Return True to suppress exceptions
- Return False or None to propagate exceptions

```python
# Exception handling in context manager
class SafeContextManager:
    def __enter__(self):
        print("Entering safe context")
        return self
    
    def __exit__(self, exc_type, exc_value, traceback):
        if exc_type is not None:
            print(f"Exception occurred: {exc_value}")
            return True  # Suppress the exception
        print("Exiting safe context normally")
        return False

# With exception (suppressed)
with SafeContextManager():
    print("Inside block")
    raise ValueError("Something went wrong!")

print("Program continues")
```

## 4. Creating a Custom Context Manager

A class can work as a context manager by defining `__enter__` and `__exit__` methods.


```python
class SimpleContextManager:
    def __enter__(self):
        print("Entering the context")
        return self

    def __exit__(self, exc_type, exc_value, traceback):
        print("Exiting the context")

with SimpleContextManager():
    print("Inside with block")
```

## Key Points

- Context managers manage resources safely.
- `with` automatically performs cleanup.
- Files are automatically closed when using `with open(...)`.
- Custom context managers use `__enter__` and `__exit__`.
- Context managers help avoid resource leaks.
