# Python Threading vs Multiprocessing


## 1. Introduction

Threading and multiprocessing are used to run multiple tasks concurrently.

They are useful when we want to improve performance or handle multiple operations at the same time.


## 2. Threading

Threading runs multiple threads inside the same process. Threads share the same memory space.

Threading is useful for I/O-bound tasks like:

- Reading files
- Calling APIs
- Database operations
- Waiting for network responses


```python
import threading
import time

def download_file(file_name):
    print(f"Starting download: {file_name}")
    time.sleep(1)
    print(f"Completed download: {file_name}")

thread1 = threading.Thread(target=download_file, args=("file1.txt",))
thread2 = threading.Thread(target=download_file, args=("file2.txt",))

thread1.start()
thread2.start()

thread1.join()
thread2.join()
```

## 3. Multiprocessing

Multiprocessing runs multiple processes. Each process has its own memory space.

Multiprocessing is useful for CPU-bound tasks like:

- Mathematical calculations
- Data processing
- Image processing
- Machine learning workloads


```python
import multiprocessing

def square(number):
    print(number * number)

process1 = multiprocessing.Process(target=square, args=(5,))
process2 = multiprocessing.Process(target=square, args=(10,))

process1.start()
process2.start()

process1.join()
process2.join()
```

## 4. ThreadPoolExecutor and ProcessPoolExecutor

### concurrent.futures Module
- Provides high-level interface for pools
- Easier to use than raw threading/multiprocessing
- Automatically manages pool of threads/processes
- Provides Future objects for async results

```python
# ThreadPoolExecutor example
from concurrent.futures import ThreadPoolExecutor
import time

def task(name, duration):
    print(f"Starting {name}")
    time.sleep(duration)
    print(f"Completed {name}")
    return f"{name} result"

# Using ThreadPoolExecutor
with ThreadPoolExecutor(max_workers=3) as executor:
    futures = [
        executor.submit(task, "Task 1", 1),
        executor.submit(task, "Task 2", 2),
        executor.submit(task, "Task 3", 1.5)
    ]
    
    for future in futures:
        print(future.result())
```

```python
# ProcessPoolExecutor example
from concurrent.futures import ProcessPoolExecutor

def cpu_task(n):
    result = sum(i * i for i in range(n))
    return result

# Using ProcessPoolExecutor
with ProcessPoolExecutor(max_workers=4) as executor:
    futures = [executor.submit(cpu_task, 1000000) for _ in range(4)]
    
    for future in futures:
        print(f"Result: {future.result()}")
```

## 5. Inter-Process Communication (IPC)

### Multiprocessing Communication
- **Queue**: Thread/process-safe queue for passing data
- **Pipe**: Two-way communication channel
- **Manager**: Shared objects (lists, dicts) across processes
- **Value/Array**: Shared memory for primitive types

```python
# Multiprocessing Queue example
from multiprocessing import Process, Queue

def worker(queue, name):
    for i in range(3):
        queue.put(f"{name} - Message {i}")

if __name__ == "__main__":
    queue = Queue()
    
    processes = [
        Process(target=worker, args=(queue, "Worker 1")),
        Process(target=worker, args=(queue, "Worker 2"))
    ]
    
    for p in processes:
        p.start()
    
    for p in processes:
        p.join()
    
    # Get all messages
    while not queue.empty():
        print(queue.get())
```

## 6. Extended Key Points & Summary

### Threading
- Multiple threads share the same memory space
- Lower overhead than multiprocessing
- Best for I/O-bound tasks (network, file, database)
- Limited by GIL for CPU-bound tasks
- Easier communication through shared memory

### Multiprocessing
- Each process has its own memory space
- Higher overhead (process creation, IPC)
- Best for CPU-bound tasks (calculations, data processing)
- Bypasses GIL limitation
- Can utilize multiple CPU cores effectively

### concurrent.futures
- ThreadPoolExecutor for thread pools
- ProcessPoolExecutor for process pools
- High-level interface for managing pools
- Future objects for async results

### Inter-Process Communication
- Queue: Thread/process-safe data passing
- Pipe: Two-way communication channel
- Manager: Shared objects across processes
- Value/Array: Shared memory for primitives

### Choosing the Right Approach
- **I/O-bound**: Use threading or asyncio
- **CPU-bound**: Use multiprocessing
- **Mixed workloads**: Combine approaches
- **Consider overhead**: Process creation is expensive
- **Consider GIL**: Threading limited for CPU tasks

## Threading vs Multiprocessing

| Feature | Threading | Multiprocessing |
| --- | --- | --- |
| Unit | Thread | Process |
| Memory | Shared memory | Separate memory |
| Best for | I/O-bound tasks | CPU-bound tasks |
| GIL impact | Affected by GIL | Avoids GIL limitation |
| Communication | Easier because memory is shared | Requires inter-process communication |
| Overhead | Lower | Higher |

## Key Points

- Use threading when tasks spend time waiting, such as network, file, or database operations.
- Use multiprocessing when tasks need heavy CPU computation, such as calculations or data processing.
- Threads share memory, processes do not.
- Multiprocessing can use multiple CPU cores effectively because it bypasses the GIL.
- GIL is an important reason to choose multiprocessing for CPU-bound Python code.
- For mixed workloads, consider combining threading, multiprocessing, or using `asyncio`.
- `ThreadPoolExecutor` and `ProcessPoolExecutor` provide convenient interfaces for managing pools.
- Always consider the overhead of creating processes vs threads.
