  5. Type Casting:

        1.   int()
        2.   float()
        3.   complex()
        4.   bool
        5.   str()

```python
# int() typecasting
num_str = "123"
num_float = 10.5

int_from_str = int(num_str)
int_from_float = int(num_float)

print(f"int('123'): {int_from_str}, Type: {type(int_from_str)}")
print(f"int(10.5): {int_from_float}, Type: {type(int_from_float)}")

# float() typecasting
num_int = 25
num_str_float = "3.14"

float_from_int = float(num_int)
float_from_str = float(num_str_float)

print(f"float(25): {float_from_int}, Type: {type(float_from_int)}")
print(f"float('3.14'): {float_from_str}, Type: {type(float_from_str)}")

# complex() typecasting
num_real = 5
num_imag = 2.5
num_str_complex = "1+2j"

complex_from_int = complex(num_real)
complex_from_parts = complex(num_real, num_imag)
complex_from_str = complex(num_str_complex)

print(f"complex(5): {complex_from_int}, Type: {type(complex_from_int)}")
print(f"complex(5, 2.5): {complex_from_parts}, Type: {type(complex_from_parts)}")
print(f"complex('1+2j'): {complex_from_str}, Type: {type(complex_from_str)}")

# bool() typecasting
int_zero = 0
int_nonzero = 100
empty_str = ""
non_empty_str = "hello"
empty_list = []
non_empty_list = [1]

bool_from_zero = bool(int_zero)
bool_from_nonzero = bool(int_nonzero)
bool_from_empty_str = bool(empty_str)
bool_from_non_empty_str = bool(non_empty_str)
bool_from_empty_list = bool(empty_list)
bool_from_non_empty_list = bool(non_empty_list)

print(f"bool(0): {bool_from_zero}, Type: {type(bool_from_zero)}")
print(f"bool(100): {bool_from_nonzero}, Type: {type(bool_from_nonzero)}")
print(f"bool(''): {bool_from_empty_str}, Type: {type(bool_from_empty_str)}")
print(f"bool('hello'): {bool_from_non_empty_str}, Type: {type(bool_from_non_empty_str)}")
print(f"bool([]): {bool_from_empty_list}, Type: {type(bool_from_empty_list)}")
print(f"bool([1]): {bool_from_non_empty_list}, Type: {type(bool_from_non_empty_list)}")

# str() typecasting
my_int_val = 456
my_float_val = 78.9
my_list_val = [10, 20]

str_from_int = str(my_int_val)
str_from_float = str(my_float_val)
str_from_list = str(my_list_val)

print(f"str(456): {str_from_int}, Type: {type(str_from_int)}")
print(f"str(78.9): {str_from_float}, Type: {type(str_from_float)}")
print(f"str([10, 20]): {str_from_list}, Type: {type(str_from_list)}")
```

# Python Version Differences: Python 2 vs Python 3

## Overview

Python 2 and Python 3 are two major versions of Python. Python 2 is older and is no longer officially supported. Python 3 is the current and recommended version for all new development.

The main reason Python 3 was introduced was to clean up old design issues and make the language more consistent.

## Key Differences

| Feature | Python 2 | Python 3 |
| --- | --- | --- |
| Print | `print "Hello"` | `print("Hello")` |
| Division | `5 / 2` gives `2` | `5 / 2` gives `2.5` |
| Strings | ASCII by default | Unicode by default |
| Input | `raw_input()` for string input | `input()` for string input |
| Range | `range()` returns a list | `range()` returns a range object |
| Support | Officially ended | Currently used and supported |

## Why Python 3 is Preferred

- It is actively maintained.
- It has better Unicode support.
- It has cleaner syntax.
- Most modern libraries support Python 3.
- New Python features are added only to Python 3.


```python
# Python 3 print function
print("Hello, Python 3!")

# Division behavior in Python 3
print(5 / 2)   # normal division
print(5 // 2)  # floor division

# range object in Python 3
numbers = range(5)
print(numbers)
print(list(numbers))
```

## Key Points

- Python 2 is outdated and should not be used for new projects.
- Python 3 uses `print()` as a function.
- Python 3 handles strings as Unicode by default.
- Division behavior is more natural in Python 3.
- For interviews and projects, Python 3 is the standard.


# Variables and Constants in Python

## What is a Variable?

A variable is a name that refers to a value stored in memory. In Python, we do not need to declare the type of a variable. Python automatically decides the type based on the assigned value.

## Creating Variables

Variables are created when we assign a value using the assignment operator `=`.


```python
name = "Kalyan"
age = 25
salary = 50000.50
is_active = True

print(name)
print(age)
print(salary)
print(is_active)

print(type(name))
print(type(age))
print(type(salary))
print(type(is_active))
```

## Variables are Dynamic

Python variables are dynamic. The same variable name can refer to different types of values at different times.


```python
x = 10
print(x, type(x))

x = "Python"
print(x, type(x))
```

## What is a Constant?

A constant is a value that should not be changed once assigned. Python does not have a special keyword to create constants. By convention, constants are written in uppercase letters.


```python
PI = 3.14159
MAX_LOGIN_ATTEMPTS = 3
DEFAULT_LANGUAGE = "Python"

print(PI)
print(MAX_LOGIN_ATTEMPTS)
print(DEFAULT_LANGUAGE)
```

## Key Points

- Variables are names used to store or refer to values.
- Python is dynamically typed.
- Constants are written in uppercase by convention.
- Python does not strictly prevent changing constants, so developers follow naming discipline.


# Variable Naming Rules and Best Practices

## Naming Rules

Python variable names must follow some rules:

- A variable name can contain letters, numbers, and underscores.
- A variable name cannot start with a number.
- A variable name cannot contain spaces.
- A variable name cannot use Python reserved keywords.
- Variable names are case-sensitive.


```python
# Valid variable names
student_name = "Ravi"
age1 = 22
_temporary_value = 100
total_marks = 450

print(student_name)
print(age1)
print(_temporary_value)
print(total_marks)
```

```python
# Invalid variable names - examples only
# 1name = "Python"       # starts with a number
# student name = "Ravi"  # contains space
# class = "Python"       # class is a reserved keyword
```

## Best Practices

- Use meaningful names.
- Use lowercase letters with underscores for variables.
- Use uppercase letters for constants.
- Avoid single-letter names except in simple loops.
- Do not use names that hide built-in functions like `list`, `str`, or `id`.


```python
# Good naming
student_total_marks = 460
average_score = 92.5
is_passed = True

# Avoid this style for normal variables
a = 460
b = 92.5
c = True

print(student_total_marks, average_score, is_passed)
```

## Key Points

- Variable names should be readable and meaningful.
- Python commonly uses `snake_case` for variable names.
- Constants are written in `UPPER_CASE`.
- Good naming makes code easier to understand and maintain.
