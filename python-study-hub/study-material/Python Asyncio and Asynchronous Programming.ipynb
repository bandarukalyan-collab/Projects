# Python Asyncio and Asynchronous Programming

## 1. Introduction to Asynchronous Programming

### What is Asynchronous Programming?
- Asynchronous programming allows a program to start a potentially long-running task and continue with other work instead of waiting for the task to complete
- Unlike synchronous programming, which executes tasks sequentially, async programming can handle multiple tasks concurrently
- Python's `asyncio` library provides a framework for writing concurrent code using the async/await syntax

### Why Use Asynchronous Programming?
- **I/O-bound tasks**: Perfect for tasks that spend time waiting (network requests, file operations, database queries)
- **Scalability**: Handle many concurrent connections with minimal resource usage
- **Responsiveness**: Keep applications responsive while waiting for I/O operations
- **Efficiency**: Better resource utilization compared to threading for I/O-heavy workloads

### Synchronous vs Asynchronous

**Synchronous (Blocking):**
```python
# Task 2 waits for Task 1 to complete
result1 = task1()
result2 = task2()
```

**Asynchronous (Non-blocking):**
```python
# Task 2 can run while Task 1 is waiting
result1 = await task1()
result2 = await task2()
```

````python
## 2. The Event Loop

### What is the Event Loop?
- The event loop is the core of asyncio
- It runs asynchronous tasks and callbacks
- It performs network I/O and runs subprocesses
- It manages the scheduling and execution of coroutines

### How the Event Loop Works
1. The event loop maintains a queue of tasks
2. It executes tasks one at a time
3. When a task awaits (e.g., waiting for I/O), the event loop switches to another task
4. This allows concurrent execution without multiple threads

### Getting the Event Loop
```python
import asyncio

# Get the running event loop
loop = asyncio.get_running_loop()

# Or create a new event loop
loop = asyncio.new_event_loop()
asyncio.set_event_loop(loop)
```
````

````python
# Example: Running asyncio
import asyncio

async def main():
    print("Hello from asyncio!")

# Run the async function
asyncio.run(main())
````

## 3. async and await Keywords

### async def
- Defines a coroutine function
- A coroutine is a special type of generator that can pause and resume execution
- When called, it returns a coroutine object, not the result

### await
- Pauses execution of the coroutine
- Yields control back to the event loop
- Waits for an awaitable to complete before resuming
- Can only be used inside async functions

````python
# Basic async/await example
import asyncio
import time

async def say_hello():
    print("Hello")
    await asyncio.sleep(1)  # Simulate I/O operation
    print("World!")

async def main():
    await say_hello()

asyncio.run(main())
````

## 4. Running Multiple Coroutines Concurrently

### asyncio.gather()
- Runs multiple awaitables concurrently
- Returns results in the same order as the inputs
- Waits for all tasks to complete
- Can handle exceptions

````python
# Running multiple coroutines concurrently
import asyncio

async def task1():
    print("Task 1 started")
    await asyncio.sleep(2)
    print("Task 1 completed")
    return "Result 1"

async def task2():
    print("Task 2 started")
    await asyncio.sleep(1)
    print("Task 2 completed")
    return "Result 2"

async def task3():
    print("Task 3 started")
    await asyncio.sleep(3)
    print("Task 3 completed")
    return "Result 3"

async def main():
    print("Starting concurrent tasks...")
    results = await asyncio.gather(task1(), task2(), task3())
    print(f"All tasks completed: {results}")

asyncio.run(main())
````

## 5. Creating Tasks

### asyncio.create_task()
- Schedules a coroutine to run soon
- Returns a Task object immediately
- Allows coroutines to run concurrently
- Can be used to fire-and-forget tasks

````python
# Creating tasks for concurrent execution
import asyncio

async def count_down(name, delay):
    for i in range(3, 0, -1):
        print(f"{name}: {i}")
        await asyncio.sleep(delay)
    print(f"{name}: Done!")

async def main():
    # Create tasks that run concurrently
    task1 = asyncio.create_task(count_down("Task A", 1))
    task2 = asyncio.create_task(count_down("Task B", 0.5))
    
    # Wait for both tasks to complete
    await task1
    await task2
    print("All tasks completed!")

asyncio.run(main())
````

## 6. asyncio.sleep()

### Purpose
- Pauses execution of a coroutine
- Non-blocking sleep (unlike time.sleep())
- Allows other coroutines to run during the delay
- Used to simulate I/O operations

````python
# asyncio.sleep vs time.sleep
import asyncio
import time

# Blocking sleep (BAD in async code)
def blocking_sleep():
    time.sleep(2)
    print("Blocking sleep done")

# Non-blocking sleep (GOOD in async code)
async def async_sleep():
    await asyncio.sleep(2)
    print("Async sleep done")

async def main():
    # This blocks everything
    # blocking_sleep()
    
    # This allows other tasks to run
    await async_sleep()

asyncio.run(main())
````

## 7. Error Handling in Asyncio

### Handling Exceptions in Coroutines
- Use try-except blocks in async functions
- Exceptions can be raised from coroutines
- asyncio.gather() can handle multiple exceptions

````python
# Error handling in asyncio
import asyncio

async def might_fail():
    await asyncio.sleep(1)
    raise ValueError("Something went wrong!")

async def main():
    try:
        await might_fail()
    except ValueError as e:
        print(f"Caught error: {e}")

asyncio.run(main())
````

## 8. Async Context Managers

### async with
- Used with context managers that support async operations
- Automatically handles cleanup of async resources
- Commonly used with async file operations, database connections

````python
# Async context manager example
import asyncio

class AsyncContextManager:
    async def __aenter__(self):
        print("Entering async context")
        return self
    
    async def __aexit__(self, exc_type, exc_val, exc_tb):
        print("Exiting async context")
        return False

async def main():
    async with AsyncContextManager():
        print("Inside async context")
        await asyncio.sleep(0.5)

asyncio.run(main())
````

## 9. Practical Example: Concurrent Network Requests

Simulating multiple API calls concurrently using asyncio.

````python
# Simulating concurrent network requests
import asyncio

async def fetch_data(url, delay):
    print(f"Fetching {url}...")
    await asyncio.sleep(delay)
    print(f"Received data from {url}")
    return f"Data from {url}"

async def main():
    urls = [
        ("https://api.example.com/users", 2),
        ("https://api.example.com/posts", 1),
        ("https://api.example.com/comments", 1.5),
    ]
    
    # Fetch all URLs concurrently
    tasks = [fetch_data(url, delay) for url, delay in urls]
    results = await asyncio.gather(*tasks)
    
    for result in results:
        print(result)

asyncio.run(main())
````

## 10. Key Points & Summary

### When to Use Asyncio
- I/O-bound tasks (network requests, file operations, database queries)
- When you need to handle many concurrent connections
- When responsiveness is important
- When you want efficient resource utilization

### When NOT to Use Asyncio
- CPU-bound tasks (use multiprocessing instead)
- When the codebase is mostly synchronous
- When simplicity is more important than performance
- When working with libraries that don't support async

### Best Practices
- Always use `asyncio.run()` to start the event loop
- Use `await` for I/O operations, not for CPU-intensive work
- Use `asyncio.gather()` for concurrent execution
- Handle exceptions properly in async functions
- Avoid mixing blocking and async code
- Use `asyncio.create_task()` for fire-and-forget operations

## 11. Async For Loops

### Async Iterators
- Use `async for` to iterate over async iterators
- Common with async generators
- Used for streaming data

````python
# Async for loop example
import asyncio

async def async_generator():
    for i in range(5):
        await asyncio.sleep(0.5)
        yield i

async def main():
    async for value in async_generator():
        print(f"Received: {value}")

asyncio.run(main())
````

## 12. Timeouts in Asyncio

### asyncio.wait_for()
- Adds a timeout to an operation
- Raises asyncio.TimeoutError if timeout is exceeded
- Useful for preventing indefinite waits

````python
# Timeout example
import asyncio

async def slow_operation():
    await asyncio.sleep(3)
    return "Operation completed"

async def main():
    try:
        result = await asyncio.wait_for(slow_operation(), timeout=2.0)
        print(result)
    except asyncio.TimeoutError:
        print("Operation timed out!")

asyncio.run(main())
````

## 13. Summary

### Key Concepts Covered:
1. **Event Loop**: Core of asyncio that manages task execution
2. **async/await**: Keywords for defining and using coroutines
3. **asyncio.gather()**: Run multiple coroutines concurrently
4. **asyncio.create_task()**: Schedule coroutines for concurrent execution
5. **asyncio.sleep()**: Non-blocking sleep for simulating I/O
6. **async with**: Async context managers for resource management
7. **async for**: Iterate over async iterators
8. **asyncio.wait_for()**: Add timeouts to operations

### Asyncio vs Other Concurrency Models:
- **Threading**: Good for I/O-bound, limited by GIL
- **Multiprocessing**: Good for CPU-bound, bypasses GIL
- **Asyncio**: Best for many I/O-bound tasks, single-threaded concurrent

### When to Choose Asyncio:
✅ Many concurrent network requests
✅ High-concurrency web servers
✅ Real-time applications
✅ When memory efficiency is important

### Common Pitfalls:
❌ Mixing blocking and async code
❌ Using time.sleep() instead of asyncio.sleep()
❌ Not handling exceptions in coroutines
❌ Forgetting to await coroutines
